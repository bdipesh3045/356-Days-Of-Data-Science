```python
from google import genai
from google.genai import types
api=""

client = genai.Client(api_key=api)

```

```python
from pydantic import BaseModel, Field
from typing import List, Optional


class Finder(BaseModel):
  recycle: bool = Field(description="In the picture weather someone is throwing stuff in dustbin or not")
```

```python
with open('driver.jpg', 'rb') as f:
    image_bytes = f.read()

response = client.models.generate_content(
  model='gemini-2.5-flash',
  contents=[
    types.Part.from_bytes(
      data=image_bytes,
      mime_type='image/jpeg',
    ),
    'You are an Ai agent who helps to determine weather in the picture someone is recycling stuff in dustbin or not is someones recyclying stuff you return true otherwise false'
  ],
  config={
  "response_mime_type": "application/json",
  "response_json_schema": Finder.model_json_schema(),
  },
)

print(response.text)
```
