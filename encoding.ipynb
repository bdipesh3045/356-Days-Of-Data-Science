```python

import kagglehub
from kagglehub import KaggleDatasetAdapter
import os



# Load the latest version
path = kagglehub.dataset_download(
  "yasserh/housing-prices-dataset",

  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)
print("Path to dataset files:", path)
print("Files inside the path:", os.listdir(path))
csv_file = os.path.join(path,os.listdir(path)[0])

```

```python
import pandas as pd
```

```python
df=pd.read_csv(csv_file)
df.head()
```

```python
df.info()
```

```python
string_columns = df.select_dtypes(include=['object'])

for i in string_columns:
  print(i)
  print(df[i].value_counts())
  print("****************")
```

```python
print(string_columns.columns)
```

```python

updated=string_columns.iloc[:,:-1]
updated
```

```python
for i in updated.columns:
  updated[i]=updated[i].map({'no': 0, 'yes': 1})

updated
```

```python
df.drop(updated.columns, axis=1,inplace=True)
```

```python
df1=df.join(updated)
```

```python
df["furnishingstatus"].value_counts()
```

```python
df.isna().value_counts()
```

```python
final=df1.join(pd.get_dummies(df["furnishingstatus"],dtype=int))
```

```python
final.drop("furnishingstatus",inplace=True,axis=1)
```
